```python
# Making a basic Bokeh line graph
# import Bokeh: Interface plotting
from bokeh.plotting import figure
from bokeh.io import output_file, show

# prepare some date
x = [1, 2, 3, 4, 5] # X axis
y = [6, 7, 8, 9, 10] # Y axis

# create a figure object
figurePlotLine = figure()
# create line plot
figurePlotLine.line(x, y)

# prepare the output
output_file('./generated/line.html')
# write the plot in the figure object
show(figurePlotLine)

# create a figure object
figurePlotTriangles = figure()
# create plot with triangles in coordinates
figurePlotTriangles.scatter(x, y, marker = 'triangle') # Use help(f.scatter) to see more options

# prepare the output
output_file('./generated/triangle.html')
# write the plot in the figure object
show(figurePlotTriangles)

# create a figure object
figurePlotCircles = figure()
# create plot with triangles in coordinates
figurePlotCircles.circle(x, y, radius = 0.05) # 0.05 dot radius. Use help(f.circle) to see more options
# but better if we use scatter

# prepare the output
output_file('./generated/circle.html')
# write the plot in the figure object
show(figurePlotCircles)
```

```python
# Making a basic Bokeh line graph
# import Bokeh: Interface plotting
# impor pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

dataCSV = pandas.read_csv('./files/data.csv')

# create a figure object
f = figure()
# create line plot
f.line(dataCSV['x'], dataCSV['y'])

# prepare the output
output_file('./generated/line_from_csv.html')
# write the plot in the figure object
show(f)
```

```python
from bokeh.plotting import figure, output_file, show

p = figure(width = 500, height = 400, tools = 'pan')
p.title.text = "Cool Data"
p.title.text_color = "Gray"
p.title.text_font = "times"
p.title.text_font_style = "bold"
p.xaxis.minor_tick_line_color = None
p.yaxis.minor_tick_line_color = None
p.xaxis.axis_label = "Date"
p.yaxis.axis_label = "Intensity"
p.line([1, 2, 3], [4, 5, 6])

output_file("./generated/graph.html")
show(p)
```

```python
from bokeh.plotting import figure, output_file, show

examplePlot = figure(width = 500, height = 400, tools = 'pan, reset')

examplePlot.title.text="Earthquakes"
examplePlot.title.text_color = "Orange"
examplePlot.title.text_font = "times"
examplePlot.title.text_font_style = "italic"

examplePlot.xaxis.minor_tick_line_color = None
examplePlot.yaxis.minor_tick_line_color = "Yellow"

examplePlot.xaxis.axis_label = "Times"
examplePlot.yaxis.axis_label = "Value"

examplePlot.scatter(x = [1,2,3,4,5], y = [5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]], color = "red", alpha = 0.5)

output_file("./files/scatter_ploting.html")
show(examplePlot)
```

```python
from bokeh.plotting import figure, output_file, show
import datetime
import pandas

externalData = pandas.read_csv('./files/adbe.csv')

externalDataPlot = figure(min_width = 500, min_height = 250, x_axis_type = 'datetime', sizing_mode="stretch_both")
externalDataPlot.line(
    [datetime.datetime.strptime(date, '%d-%b-%y') for date in externalData['Date']], 
    externalData['Close'], 
    color = 'Orange', 
    alpha = 0.5
)

output_file("./generated/time_series.html")
show(externalDataPlot)
```
