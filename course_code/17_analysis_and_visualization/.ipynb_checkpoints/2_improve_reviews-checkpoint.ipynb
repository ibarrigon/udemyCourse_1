```python
from datetime import datetime
from pytz import utc
import pandas, matplotlib.pyplot as plt # normally renamed to plt

# load file
reviews_data = pandas.read_csv('./files/reviews.csv', parse_dates = ['Timestamp'])
reviews_data.head()
```

```python
reviews_data['Day'] = reviews_data['Timestamp'].dt.date
reviews_data
```

```python
reviews_data_average = reviews_data.groupby(['Day']).mean(['Rating'])
reviews_data.head()
```

```python
plt.plot(reviews_data_average.index, reviews_data_average['Rating'])
```

```python
int(reviews_data_average['Rating'].min())
```

```python
plt.figure(figsize = (25, 3))
plt.plot(reviews_data_average.index, reviews_data_average['Rating'])
```

### Average rating by month

```python
reviews_data['Month'] = reviews_data['Timestamp'].dt.strftime('%Y-%m')
reviews_month_average = reviews_data.groupby(['Month']).mean(['Rating'])
plt.figure(figsize = (25, 3))
plt.plot(reviews_month_average.index, reviews_month_average['Rating'])
```

### Average rating by month by course

```python
reviews_data['Month'] = reviews_data['Timestamp'].dt.strftime('%Y-%m') # Just for consistency
by_course = reviews_data.groupby(['Month', 'Course Name']).mean(['Rating'])
by_course[:20]
```

```python
by_course = reviews_data.groupby(['Month', 'Course Name']).mean(['Rating']).unstack()
by_course[:20]
```

```python
by_course[-20:]
```

```python
by_course.plot(figsize = (25, 8))
```

```python
by_course_2 = reviews_data.groupby(['Month', 'Course Name']).count().unstack()
by_course_2.plot(figsize = (25, 8))
```

```python
by_course_3 = reviews_data.groupby(['Month', 'Course Name'])['Rating'].count().unstack()
by_course_3.plot(figsize = (25, 8))
```

### What day people give best rating?

```python
reviews_data['Weekday'] = reviews_data['Timestamp'].dt.strftime('%A')
weekday_average = reviews_data.groupby(['Weekday']).mean(['Rating'])
plt.plot(weekday_average.index, weekday_average['Rating'])
```

```python
weekday_average = weekday_average.sort_values('Weekday')
plt.plot(weekday_average.index, weekday_average['Rating'])
```

```python
reviews_data['Daynumber'] = reviews_data['Timestamp'].dt.strftime('%w')
reviews_data
```

```python
weekday_average = reviews_data.groupby(['Weekday', 'Daynumber']).mean(['Rating'])
weekday_average = weekday_average.sort_values('Daynumber')
plt.figure(figsize = (10, 3))
plt.plot(weekday_average.index.get_level_values(0), weekday_average['Rating'])
```
